```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def read_files(file_paths):
    texts = []
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as file:
            texts.append(file.read())
    return texts
```

```python
def compute_similarity(file_paths):
    texts = read_files(file_paths)
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(texts)
    similarity_matrix = cosine_similarity(tfidf_matrix)

    results = []
    for i in range(len(file_paths)):
        for j in range(i + 1, len(file_paths)):
            similarity_score = similarity_matrix[i, j]
            results.append((file_paths[i], file_paths[j], similarity_score))
    return results
```

```python
def main():
    # Define the list of file paths
    file_paths = ['/content/drive/MyDrive/john.txt', '/content/drive/MyDrive/juma.txt', '/content/drive/MyDrive/fatma.txt']

    # Compute similarity
    similarity_results = compute_similarity(file_paths)

    # Print the results
    for result in similarity_results:
        print(result)
```

```python
if __name__ == "__main__":
    main()
```
